```python
# Import relevant packages and CSV file
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt
med_data = pd.read_csv('/Users/astro/Downloads/medical_clean.csv')
%matplotlib inline
df = med_data[['Anxiety','ReAdmis']]

df.head()
```

```python
contingency= pd.crosstab(df['Anxiety'], df['ReAdmis'])
contingency
```

```python
contingency_pct = pd.crosstab(df['Anxiety'], df['ReAdmis'], normalize='index')
contingency_pct
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(contingency, annot=True, cmap="YlGnBu")
```

```python
# Chi-Square test of independence
c, p, dof, expected = chi2_contingency(contingency)
# Print the P-Value
print(p)
```
